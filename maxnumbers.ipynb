```python
def find_maximum (numbers):
    if not numbers:
        raise ValueError("Cannot find max of emptpy list")
    max_value = numbers[0]
    for num in numbers:
        if num > max_value:
            max_value = numbers
    return max_value

print(find_maximum([3, 1, 4, 1, 5, 9, 2, 6, 5, 3, 5]))  
```
